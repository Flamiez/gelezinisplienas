```python
import os
```

```python
images_folder = "../Datasets/green/GeneratedImages"
masks_folder = "../Datasets/green/GeneratedMasks"
```

```python
startNumber = 0
```

```python
image_files = sorted(
    os.listdir(images_folder),
    key=lambda x: int(os.path.splitext(x)[0])
)

for img_name in image_files:
    name, ext = os.path.splitext(img_name)
    temp_name = f"{name}_tmp{ext}"

    # rename image to temp
    os.rename(
        os.path.join(images_folder, img_name),
        os.path.join(images_folder, temp_name)
    )

    # rename mask to temp
    mask_path = os.path.join(masks_folder, img_name)
    mask_temp = os.path.join(masks_folder, temp_name)
    if os.path.exists(mask_path):
        os.rename(mask_path, mask_temp)
    else:
        print(f"Mask missing for {img_name}")

temp_files = sorted(
    [f for f in os.listdir(images_folder) if f.endswith("_tmp.png") or f.endswith("_tmp.jpg")],
    key=lambda x: int(x.split("_tmp")[0])  # sort by original number
)

for i, temp_name in enumerate(temp_files):
    ext = os.path.splitext(temp_name)[1]
    new_name = f"{startNumber + i}{ext}"

    # rename image temp → final
    os.rename(
        os.path.join(images_folder, temp_name),
        os.path.join(images_folder, new_name)
    )

    # rename mask temp → final
    mask_temp = os.path.join(masks_folder, temp_name)
    mask_new = os.path.join(masks_folder, new_name)
    os.rename(mask_temp, mask_new)

print("Renaming completed successfully!")
```

Deleting masks that don't exist in images

```python
images = set(os.listdir(images_folder))
masks = set(os.listdir(masks_folder))
orphan_masks = masks - images

print(f"Found {len(orphan_masks)} orphaned masks.")

for mask_file in orphan_masks:
    mask_path = os.path.join(masks_folder, mask_file)
    print(f"Deleting mask: {mask_file}")
    os.remove(mask_path)

print("Done. All orphan masks removed.")
```
